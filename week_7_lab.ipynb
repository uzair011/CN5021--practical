# **Line Chart**

```python
import matplotlib.pyplot as plt

X = [ "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
y = [ 22, 20, -1, 15, -10, 11, 14 ]
y1 = [ 21, 16, -3, 14, -10, 8, 15 ]

plt.plot(X,y) # creating/ plotting the data - table

plt.title(" Temperature vs week")
plt.xlabel("Day")
plt.ylabel("Temperature")

plt.plot(X, y, "r", linewidth=4)  # formatting the line with colours
plt.plot(X, y, ">")  # formatting the line type

plt.savefig("tempVSday", format="pdf") # save in pdf file

plt.annotate(xy = ['Friday', -10], text= 'Lowest') # annotate

plt.plot(X, y1, "b", linestyle="dashed")  # formatting the line with colours
plt.plot(X, y1, "*")  # formatting the line type

# change the background
plt.style.use("dark_background")
#plt.style.use("fast")
#plt.style.use("grayscale")
#plt.style.use("bmh")

plt.legend(["London", "London_temp", "England", "England_temp" ], loc=0)
```

# **Bar Chart**

```python
# 2D

import matplotlib.pyplot as plt

X1 = [1, 3, 5, 7, 9]
y1 = [5, 3, 7, 9, 2]
X2 = [3, 8, 9, 3, 1]
y2 = [9, 7, 4, 9, 6]

plt.bar(X1, y1, label="X1 vs Y1", color = "g" )
plt.bar(X2, y2, label="X2 vs Y2", color = "b" )

plt.legend()
plt.xlabel("X value")
plt.ylabel("y value")

plt.title("X vs Y values...")

plt.show()
```

```python
# 2D - horizontal

import matplotlib.pyplot as plt

X1 = [1, 3, 5, 7, 9]
y1 = [5, 3, 7, 9, 2]
X2 = [3, 8, 9, 3, 1]
y2 = [9, 7, 4, 9, 6]

plt.barh(X1, y1, label="X1 vs Y1", color = "g" )
plt.barh(X2, y2, label="X2 vs Y2", color = "b" )

plt.legend()
plt.xlabel("X value")
plt.ylabel("y value")

plt.title("X vs Y values...")

plt.show()
```

# **Sub Plotting**

```python
import matplotlib.pyplot as plt

data = {
    "java": 40,
    "python": 35,
    "c++": 23,
    "C": 34
}
names = list(data.keys())
values = list(data.values())


fig, axs = plt.subplots(1, 3, figsize = (9, 3))
axs[0].bar(names, values)
axs[1].scatter(names, values)
axs[2].plot(names, values)

fig.suptitle("Categorical plotting")
```

# **Pie Chart**

```python
import matplotlib.pyplot as plt

data = {
    "java": 40,
    "python": 35,
    "c++": 23,
    "C": 34
}
names = list(data.keys())
values = list(data.values())

# explode a slice - ie: c++
explode = (0, 0, 0.1, 0)

fig, axs = plt.subplots()

axs.pie(values, explode=explode, labels=names, shadow=True, startangle=180)

plt.show()
```

# **Histogram**

```python
import numpy as np
import matplotlib.pyplot as plt

mu, sigma = 0, 0.1
data = np.random.normal(mu, sigma, 1000)

fig = plt.figure()
ax = fig.add_subplot(111)

numBins = 15
ax.hist(data, numBins, color="red")
plt.xlabel("Data")
plt.ylabel("Frequency")
plt.show()
```
