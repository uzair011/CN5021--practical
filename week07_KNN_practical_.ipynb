```python
# imnport necessary python libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
# read the data using pandas
df = pd.read_csv("/content/heart.csv")
```

```python
# print...
df.head(7)
```

```python
y = df["target"].values
X = df.drop("target", axis = 1).values   # drop the target column to save all the other data as x

print(y)
print(X)
```

```python
from sklearn.model_selection import train_test_split

# split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier

# innitialize KNN
model_0 = KNeighborsClassifier(n_neighbors = 5)

# train the model
model_0.fit(X_train, y_train)

# check the accuracy of the model
accuracy = model_0.score(X_test, y_test)

print ("Accuracy of the model over the test set is: ", accuracy)
```

```python
from sklearn.neighbors import KNeighborsClassifier

# initialize the KNN  - 2
model_1 = KNeighborsClassifier(n_neighbors = 12)

# train the model
model_1.fit(X_train, y_train)

# check the accuracy
accuracy_1 = model_1.score(X_test, y_test)

print("Accuracy of the model_1 is : ", accuracy_1)
```

```python
# plotting

my_neighbours = [ 5, 10 ]
my_accuracy = [accuracy,accuracy_1]

plt.title("The chart shows no. of neighbours vs test accuracy")
plt.plot(my_neighbours, my_accuracy)
plt.show()
```
