```python
# Interactive while loop

sum = 0
count = 0
average = 0

while(True):
  input1 = float(input("Enter an integer: "))
  count += 1;
  sum = sum + input1;
  input2 = input("Do you want to continue entering a new number? (Y/N)")

  if (input2 == "Y" or input2 == "y"):
    continue;
  else:
    break;

  average = sum/count;

print("Total values entered are :", count)
print("Total sum is: ", sum)
print("Average is: ", average)

```

```python
# Nested Loops
for i in range(10):
  for j in range(i):
    print(i, end="")
  print("\n")
```
